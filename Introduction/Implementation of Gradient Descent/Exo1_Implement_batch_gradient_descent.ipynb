```python
import numpy as np

# Features data
X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])

# Target data
y = np.array([[1], [2], [3], [4]])

# Save the features and target data to files
np.savetxt('features.txt', X)
np.savetxt('target.txt', y)
```

```python
import numpy as np

#linear regression defined as h(x) = y = theta0 + theta1 * x1 
def linear_regression(X, y, learning_rate=0.01, num_epochs=100):
    """
    Train a linear regression model using batch gradient descent
    :param X: Features data
    :param y: Target data
    :param learning_rate: Learning rate for gradient descent
    :param num_epochs: Number of epochs
    :return: Trained model parameters
    """
    m, n = X.shape #number of samples and number of features
    theta = np.zeros((n, 1)) #define theta0 and theta1 as 0
    print("array")
    
    for epoch in range(num_epochs):
        y_pred = X @ theta # compute the predicted values through a matrix multiplication
        error = y - y_pred # compute the error
        gradient = -(2/m) * X.T @ error #derivative of the cost function
        theta = theta - learning_rate * gradient # update the model parameters
        print(theta)

    return theta

# Load the features and target data
X = np.loadtxt('features.txt')
y = np.loadtxt('target.txt').reshape(-1, 1)

# Train the linear regression model using batch gradient descent
theta = linear_regression(X, y)
```
